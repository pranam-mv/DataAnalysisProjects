13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
corr = df['view_count'].corr(df['dislikes'])

print(f"Pearson correlation coefficient between views and dislikes: {corr}")

sns.scatterplot(x='view_count', y='dislikes', data=df)
```

```python
# The corelation between the views and dislikes is 68 %. It means there is moderate corerelation between these two variables.
# We can conculde more number of views will the effect the number of video dislikes but only for some extent
```
